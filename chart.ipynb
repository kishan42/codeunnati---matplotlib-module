```python
import matplotlib.pyplot as plt 
import numpy as np

fig,ax = plt.subplots()

x = np.linspace(0,20,1000)
ax.plot(x,np.cos(x))
ax.axis('equal')

ax.annotate("local maximum", xy = (6.28,1) , xytext = (10,4), arrowprops=dict(facecolor='black',shrink=0.05))
ax.annotate('local minimum', xy = (5*np.pi,-1) , xytext = (2,-6),arrowprops=dict(arrowstyle="->",connectionstyle="angle3,angleA=0,angleB=-80"))
```

```python
import matplotlib.pyplot as plt 
import numpy as np

x = np.arange(0,10,0.1)
y = np.sin(x)  

plt.plot(x,y)
plt.axis('equal')    # it is use for same space of upper and lower side


plt.annotate("local maximum", xy = (np.pi/2,1) , xytext = (7,2.5), arrowprops=dict(facecolor='black',shrink=0.05))    # here, xy = (highvalue(x),highvalue(y)) in sin graph  xytext = (starting text according to x-axies,starting text according to y-axies)
plt.annotate('local minimum', xy = (3*np.pi/2,-1) , xytext = (1,-3),arrowprops=dict(arrowstyle="->",connectionstyle="angle3,angleA=8,angleB=-60"))  # here, xy = (lowalue(x),lowvalue(y)) in sin graph 
```

```python
import matplotlib.pyplot as plt
w =4
h =3
d =70

plt.figure(figsize=(w,h),dpi =d)   #dpi is use of zoom out of graph
plt.axis([0,5,0,5])
x = [0,3,5]
y = [1,4,3.5]
label_x= 1
label_y= 4
arrow_x= 3
arrow_y= 4

arrow_properties = dict(facecolor="black",width=1,headwidth=7,shrink=0.1)

plt.annotate("maximum",xy=(arrow_x,arrow_y),xytext=(label_x,label_y), arrowprops = arrow_properties)
plt.plot(x,y)
```

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()

ax = plt.axes(projection ='3d')

z = np.linspace(0,1,100)
x = z*np.sin(25*z)
y = z*np.cos(25*z)

ax.plot3D(x,y,z,'green')
ax.set_title('3D line plot')
plt.show()
```
